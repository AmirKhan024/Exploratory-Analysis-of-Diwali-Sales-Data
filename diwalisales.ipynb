```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)  # axis =0 :- to delete rows # axis=1 :- to delete the column
```

```python
df.info()
```

```python
pd.isnull(df).sum()  # basically it will show how many null values are there in each columns
```

```python
df.shape
```

```python
df.dropna(inplace=True)  # basically this line will drop all null values
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']=df['Amount'].astype("int")   # since amount dtype was float, we used astype to convert float to int
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Cust_name':"CustomerName"})  # we have not used inplace or assigned it to new df, so it is not saved
```

```python
df[['Age','Orders','Amount']].describe()  # apply describe method on specific cols
```

```python
# this code snippet basically tells that how much orders had been placed by female and male
cnt=sns.countplot(x='Gender', data=df)
for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
# this snippet tells that how much amount is spent by female and male
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot( x= 'Gender', y='Amount', data=sales_gender)
```

```python
cnt = sns.countplot(x = "Age Group" , hue = "Gender" , data=df)
for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot( x= 'Age Group', y='Amount', data=sales_age)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})  # used to specify appearance of figure, 16 is widht and 5 is height and rc stands for run command

sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state, x= 'State', y='Amount')
```

```python
cnt = sns.countplot(x='Marital_Status' , data=df)
sns.set(rc={'figure.figsize':(6,5)})
for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
sales_maritalstatus=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_maritalstatus, x= 'Marital_Status', y='Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
cnt = sns.countplot(x='Occupation' , data=df)

for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occupation, x= 'Occupation', y='Amount')
```

```python
sns.set(rc={'figure.figsize':(16,5)})
plt.xticks(rotation=45)
cnt = sns.countplot(x='Product_Category' , data=df)

for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(16,5)})
plt.xticks(rotation=45)
sns.barplot(data=sales_product, x= 'Product_Category', y='Amount')
```

```python
sales_productid=df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_productid, x= 'Product_ID', y='Amount')
```
